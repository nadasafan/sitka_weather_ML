```python
import csv
file_name = 'sitka_weather_2014.csv'
with open(file_name) as f:
    reader = csv.reader(f)
    headers = next(reader)
    print(headers)
    for index, header in enumerate(headers):
        print(index, header)
    # reading max temperatures
    high_temp = []
    for line in reader:
        high_temp.append(int(line[1]))
    print(high_temp)
       
```

```python
#visualizing data
import matplotlib.pyplot as plt
plt.plot(high_temp, c='red')
```

```python
#spllittind data
import numpy as np
indices = range(0,len(high_temp))
training_indecies = np.random.choice(high_temp,100,replace = False)
remaining_indices = list(set(indices)-set(training_indecies))
val_indices = np.random.choice(remaining_indices,100,replace = False)
test_indices = set(remaining_indices) - set(val_indices)
```
